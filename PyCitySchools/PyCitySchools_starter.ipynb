```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools=len(school_data)
total_students=len(student_data)
total_budget=school_data["budget"].sum()
avg_math_score=school_data_complete["math_score"].sum()/total_students
avg_reading_score=school_data_complete["reading_score"].sum()/total_students
math_pass = school_data_complete[school_data_complete["math_score"] >=70]
percent_math_pass = 100*math_pass["math_score"].count()/total_students
reading_pass = school_data_complete[school_data_complete["reading_score"] >=70]
percent_reading_pass = 100*reading_pass["reading_score"].count()/total_students
overall_pass = reading_pass[reading_pass["math_score"] >= 70]
percent_overall_pass=100*overall_pass["reading_score"].count()/total_students
percent_math_pass
```

```python
dist_summary=pd.DataFrame({
    'Total Schools' : [total_schools],
    "Total Students" : [total_students],
    "Total Budget"  : [total_budget],
    "Average Math Score" : [avg_math_score],
    "Average Reading Score" :[avg_reading_score],
    "% Passing Math" :  [percent_math_pass],
    "% Passing Reading" : [percent_reading_pass],
    "% Overall Passing" : [percent_overall_pass]
})
dist_summary.style.format({"Total Students" :"{0:,}",
                           "Total Budget":"${0:,.2f}",
                           "Average Math Score":"{0:,.6f}"
                          })
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary=school_data[["school_name", "type", "size" , "budget"]]
school_summary["Per Student Budget"] = school_summary["budget"]/school_summary["size"]
school_df=school_data_complete.groupby(["school_name"])
math_average=school_df["math_score"].mean()
reading_average=school_df["reading_score"].mean()
school_summary = pd.merge(school_summary, math_average, how="left", on=["school_name", "school_name"])
school_summary = pd.merge(school_summary, reading_average, how="left", on=["school_name", "school_name"])
#Unable to use school_df here since error doing comparison operations with group-by data
#has to go back to the school_data_complete
math_pass=school_data_complete[school_data_complete["math_score"] >=70]
reading_pass=school_data_complete[school_data_complete["reading_score"] >=70]
overall_pass=math_pass[math_pass["reading_score"] >=70]
grouped_math=math_pass.groupby("school_name")["student_name"].count().reset_index()
grouped_math.columns=["school_name", "math_count"]
grouped_reading=reading_pass.groupby("school_name")["student_name"].count().reset_index()
grouped_reading.columns=["school_name", "reading_count"]
grouped_overall=overall_pass.groupby("school_name")["student_name"].count().reset_index()
grouped_overall.columns=["school_name", "overall_count"]
```

```python



#Add to the dataframe
school_summary = pd.merge(school_summary, grouped_math, how="left", on=["school_name", "school_name"])
school_summary = pd.merge(school_summary, grouped_reading, how="left", on=["school_name", "school_name"])
school_summary = pd.merge(school_summary, grouped_overall, how="left", on=["school_name", "school_name"])

school_summary["% Passing Math"]= 100*school_summary["math_count"]/school_summary["size"]
school_summary["% Passing Reading"]= 100*school_summary["reading_count"]/school_summary["size"]
school_summary["% Overall Passing"] = 100*school_summary["overall_count"]/school_summary["size"]
school_summary.drop("math_count", axis=1, inplace=True)
school_summary.drop("reading_count", axis=1, inplace=True)
school_summary.drop("overall_count", axis=1, inplace=True)
school_summary
```

```python

school_summary.rename(columns={"school_name":"School Name",
                               "type":"School Type",
                               "size" : "Total Students",
                               "budget" :"Total School Budget",
                               "math_score" : "Average Math Score",
                               "reading_score": "Average Reading Score"
                              })
school_summary.style.format({"Total Budget":"${0:,.2f}",
                             "Average Math Score":"{0:,.6f}",
                             "Per Student Budget" : "${0:,}"
                             
                          })
school_summary.set_index("school_name")
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
school_summary=school_summary.set_index("school_name")
school_summary.sort_values(by=["% Overall Passing"], inplace=True, ascending=False)
school_summary.head()
```
